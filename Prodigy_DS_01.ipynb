```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Plotting Bar Graph
plt.figure(figsize=(22, 6))
sns.countplot(x='AREA NAME', data=df, palette='viridis')
plt.title('Bar Graph of AREA vs CRIME')
plt.xlabel('AREA')
plt.ylabel('CRIME')
plt.show()
```

```python
#plotting histogram
plt.figure(figsize =(10,5))
sns.histplot(data = df, x = 'Crm Cd',bins = np.arange(0,1100,100),color = 'skyblue')
plt.show()
```
